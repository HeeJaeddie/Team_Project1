```python
import pandas as pd

file = 'Final_flight_data.csv'
df = pd.read_csv(file, encoding='cp949')
df.head()
```

```python
# 출발 공항과 도착 공항으로 나누기
df[['departure', 'arrival']] = df['airport'].str.split('-', expand=True)
df.head()
```

```python
# 'airport' 컬럼 삭제
df = df.drop(columns='airport')

df.head()
```

```python
# 'day' 컬럼 삭제
df = df.drop(columns='day')

df.info()
```

```python
df.head()
```

```python
# 'leavetime','reachtime' 컬럼 'HH:MM' 포맷으로 바꾸기
df['leavetime'] = (df['leavetime'] // 100).astype(str).str.zfill(2) + ':' + (df['leavetime'] % 100).astype(str).str.zfill(2)
df['reachtime'] = (df['reachtime'] // 100).astype(str).str.zfill(2) + ':' + (df['reachtime'] % 100).astype(str).str.zfill(2)

# 출발 날짜 시각, 도착 날짜 시각 컬럼 만들기
df['departure_datetime'] = pd.to_datetime(df['date'] + ' ' + df['leavetime'], format='%Y-%m-%d %H:%M')
df['arrival_datetime'] = pd.to_datetime(df['date'] + ' ' + df['reachtime'], format='%Y-%m-%d %H:%M')

df.head()
```

```python
# 'leavetime', 'reachtime', 'date' 컬럼 삭제
df = df.drop(columns=['leavetime', 'reachtime', 'date'])
df.head()
```

```python
# 제주 기준 확인용 컬럼 생성
df['check'] = df.apply(lambda row: 'back' if row['departure'] == 'CJU' else ('come' if row['arrival'] == 'CJU' else 'none'), axis=1)

df.head()
```

```python
# 성인 요금 컬럼명 변경
df.rename(columns={'charge': 'adult_charge'}, inplace=True)

# 소아 요금 컬럼 생성
df['child_charge'] = df['adult_charge'] * 0.75

df.head()
```

```python
# 소수점 제거
df['child_charge'] = df['child_charge'].astype(int)
df.info()
```

```python
airport_dict = {
    'GMP': '김포',
    'PUS': '부산',
    'KWJ': '광주',
    'MWX': '무안',
    'TAE': '대구',
    'RSU': '여수',
    'USN': '울산',
    'WJU': '원주',
    'CJJ': '청주',
    'KPO': '포항',
    'CJU' :'제주'
}

# 한국어로 공항명 컬럼 생성
df['departure_kor'] = df['departure'].map(airport_dict)
df['arrival_kor'] = df['arrival'].map(airport_dict)

df.head()
```

```python
df.to_csv('preprocessed_flight2.csv', index=False, encoding='cp949')
```
